```python
imdb_general = pd.read_csv('title.akas.tsv', sep='\t', low_memory=False)
imdb_basics = pd.read_csv('title.basics.tsv', sep='\t', low_memory=False)
imdb_ratings = pd.read_csv('title.ratings.tsv', sep='\t', low_memory=False)
imdb_names = pd.read_csv('name.basics.tsv', sep='\t', low_memory=False)
imdb_crew = pd.read_csv('title.crew.tsv', sep='\t', low_memory=False)
```

```python
imdb_brasil = imdb_general.query("region == 'BR'")
```

```python
imdb_brasil = imdb_brasil.rename({'titleId': 'tconst'}, axis=1)
```

```python
imdb_brasil_basics = imdb_brasil.merge(imdb_basics, on='tconst', how='inner')
```

```python
imdb_brasil_basics_ratings = imdb_brasil_basics.merge(imdb_ratings, on='tconst', how='inner')
```

```python
imdb_names = imdb_names.rename({'nconst': 'directors'}, axis=1)
```

```python
imdb_directors = imdb_names.merge(imdb_crew, on='directors', how='inner')
```

```python
imdb_brasil_basics_ratings_directors = imdb_brasil_basics_ratings.merge(imdb_directors, on='tconst', how='inner')
```

```python
imdb_brasil_basics_ratings_directors = imdb_brasil_basics_ratings_directors[['title',
                                                                             'originalTitle',
                                                                             'titleType',
                                                                             'startYear',
                                                                             'primaryName',
                                                                             'genres',
                                                                             'runtimeMinutes', 
                                                                             'averageRating',
                                                                             'numVotes']]
```

```python
imdb_brasil_basics_ratings_directors
```

```python
imdb_brasil_movies = imdb_brasil_basics_ratings_directors.query('titleType == "movie"')
```

```python
imdb_brasil_good_movies = imdb_brasil_movies.query("averageRating > 6.9")
```

```python
imdb_brasil_good_movies_votes = imdb_brasil_good_movies.query('numVotes > 999')
```

```python
imdb_brasil_good_movies_votes
```

```python
imdb_brasil_limpeza = imdb_brasil_good_movies_votes[~imdb_brasil_good_movies_votes.runtimeMinutes.str.contains('N')]
```

```python
imdb_brasil_limpeza.drop_duplicates(subset ="originalTitle",
                     keep = 'first', inplace = True)
```

```python
imdb_brasil_limpeza
```

```python
imdb_brasil_limpeza.reset_index(drop=True)
```

```python
imdb_brasil_limpeza[["genre", "genre1", "genre2"]] = imdb_brasil_limpeza["genres"].str.split(pat=",", expand=True)
```

```python
imdb_brasil_limpeza["genre"].value_counts()
```

```python
imdb_brasil_selecao = imdb_brasil_limpeza[['title',
                                         'originalTitle',
                                         'primaryName',
                                         'startYear',
                                         'genre',
                                         'runtimeMinutes', 
                                         'averageRating']]
```

```python
imdb_brasil_selecao
```

```python
imdb_brasil_selecao['genre'].unique()
```

```python
imdb_brasil_selecao = imdb_brasil_selecao.replace({'genre' : {'Animation':'Animação',
                                                             'Action':'Ação',
                                                             'Biography':'Biografia',
                                                             'Adventure':'Aventura',
                                                             'Documentary':'Documentário',
                                                             'Comedy':'Comédia',
                                                             'Mystery':'Mistério',
                                                             'Fantasy':'Fantasia',
                                                             'Film-Noir':'Noir',
                                                             'History':'História',}}) 
```

```python
imdb_brasil_selecao['genre'].unique()
```

```python
imdb_brasil_selecao_nota = imdb_brasil_selecao.sort_values(by=['averageRating'], ascending=False)
```

```python
imdb_brasil_selecao_nota.reset_index(drop=True)
```

```python
imdb_brasil_selecao_nota.dtypes
```

```python
imdb_brasil_selecao_nota.runtimeMinutes = imdb_brasil_selecao_nota.runtimeMinutes.astype(int)
```

```python
imdb_brasil_selecao_nota = imdb_brasil_selecao_nota.rename({'primaryName': 'director'}, axis=1)
imdb_brasil_selecao_nota = imdb_brasil_selecao_nota.rename({'startYear': 'year'}, axis=1)
```
